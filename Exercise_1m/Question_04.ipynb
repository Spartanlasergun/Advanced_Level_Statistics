# Question Four

(a)

| length(cm) | frequency | cum. frequency |
| --- | --- | --- |
| 0-20 | 20 | 20 |
| 20-30 | 20 | 40 |
| 30-40 | 25 | 65 |
| 40-50 | 4 | 69 |

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

y = [0, 20, 40, 65, 69]
x = [0, 10, 25, 35, 45]

f = interp1d(x, y, kind='quadratic')

x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.scatter(x, y)
plt.plot(x_new, y_new)
plt.show()
```

```python
x_list = x_new.tolist()
y_list = y_new.tolist()

for item in y_list:
    if item <= 34.5:
        median_index = y_list.index(item)
median = x_list[median_index]
print(median)
```

```python
for item in y_list:
    if item <= 51.75:
        upper_index = y_list.index(item)
upper_quartile = x_list[upper_index]
print(upper_quartile)
```

```python
for item in y_list:
    if item <= 17.25:
        lower_index = y_list.index(item)
lower_quartile = x_list[lower_index]
print(lower_quartile)
```

```python
import seaborn as sns

data = [0, lower_quartile, median, upper_quartile, 69]
sns.boxplot(data, whis=69, orient='h')
```

```python
Q3_Q2 = upper_quartile - median
Q2_Q1 = median - lower_quartile

print("Q3 - Q2 = " + str(Q3_Q2))
print("Q2 - Q1 = " + str(Q2_Q1))
print("The distribution is negatively skewed")
```
