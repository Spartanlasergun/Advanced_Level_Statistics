```python
import seaborn as sns
import pandas as pd

group_one = [26, 26, 24, 24, 25, 22, 22, 23, 23, 23, 23, 20, 20, 20, 21, 21,
             18, 16, 16, 17, 14, 14, 15]

group_two = [24, 25, 22, 22, 22, 23, 23, 20, 20, 21, 18, 19, 19, 19, 19, 16, 16,
             17, 17, 14, 14, 12, 9]

df = pd.DataFrame(list(zip(group_one, group_two)), columns =['group one', 'group two'])

sns.boxplot(data=df, whis=26, orient='h')
```

The outliers in group two has made the data neagtively skewed, whereas group one appears to be a more balanced distribution.
