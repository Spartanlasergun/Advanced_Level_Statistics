```python
import statistics

data = [1, 22, 30, 19, 15, 29,
       45, 53, 2, 5, 31, 47,
       17, 48, 29, 9, 29, 2,
       10, 12, 25, 10, 13, 19]

quantiles = statistics.quantiles(data, n=4)
median = quantiles[1]
upper_quartile = quantiles[2]
lower_quartile = quantiles[0]

print("median = " + str(median))
print("upper quartile = " + str(upper_quartile))
print("lower quartile = " + str(lower_quartile))
```

```python
import seaborn as sns

box_data = [min(data), lower_quartile, median, upper_quartile, max(data)]
sns.boxplot(x=box_data, whis=53)
```

The data is positively skewed
