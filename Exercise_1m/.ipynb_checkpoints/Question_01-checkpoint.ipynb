# Question One

| Length of call (min) | Number of Calls | Cum. Frequency |
| --- | --- | --- |
| &le; 1 | 8 | 8 |
| 1-2 | 11 | 19 |
| 2-3 | 17 | 36 |
| 3-5 | 8 | 44 |
| 5-10 | 6 | 50 |
| &ge; 10 | 0 | 50 |

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

y = [0, 8, 19, 36, 44, 50]
x = [0, 1, 2, 3, 5, 10]

interpolation_function = interp1d(x, y, kind='quadratic') # create a function of y and x

x_new = np.linspace(min(x), max(x), 100)  # generate more x values for a smooter curve
y_new = interpolation_function(x_new)     # generate the corresponding y-values from the original function

plt.plot(x_new, y_new)
plt.scatter(x, y)
plt.show()
```

```python
x_list = x_new.tolist()
y_list = y_new.tolist()

for item in y_list:
    if item <= 25:
        median_index = y_list.index(item)
median = x_list[median_index]

for item in y_list:
    if item <= 37.5:
        upper_index = y_list.index(item)
upper_quartile = x_list[upper_index]

for item in y_list:
    if item <= 12.5:
        lower_index = y_list.index(item)
lower_quartile = x_list[lower_index]

print("(b)")
print("median = " + str(median))
print("upper quartile = " + str(upper_quartile))
print("lower quartile = " + str(lower_quartile))
```

```python
import seaborn as sns

data = [0, lower_quartile, median, upper_quartile, 10]

# Note: by default seaborn will not plot the upper or lower limits of the data it is found to be greater in value
# than the interquartile range. Hence, 'whis' must be set to the full range of the data for a proper plot.
sns.boxplot(x=data, whis=10) 
```

The data in the above plot is positively skewed
