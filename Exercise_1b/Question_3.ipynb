```python
import matplotlib.pyplot as plt

frequency = [62, 70, 88, 280, 30]
mid_points = [15, 40, 70, 120, 225]
raw_data =[]
index = 0
for item in frequency:
    count = item
    while count != 0:
        raw_data.append(mid_points[index])
        count = count - 1
    index = index + 1

plt.hist(raw_data, bins=[0, 25, 60, 80, 150, 300], edgecolor="black")
```
