```python
import matplotlib.pyplot as plt
import math

time = [5, 10, 20, 25, 40, 45]
frequency = [2, 12, 7, 15, 2, 0]

frequency_density = [0.4, 1.2, 1.4, 1, 0.4]

n = sum(frequency)
data_range = 40
number_of_intervals = math.sqrt(n)
interval_width = int(data_range/number_of_intervals)

plt.hist(frequency_density, bins=interval_width)
plt.show()
```
