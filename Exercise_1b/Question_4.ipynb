```python
import matplotlib.pyplot as plt

frequency = [21, 62, 55, 50, 12]
interval_midpoints = [45, 53, 58, 65, 73]
raw_data = []
index = 0
for item in frequency:
    count = item
    while count != 0:
        raw_data.append(interval_midpoints[index])
        count = count - 1
    index = index + 1

plt.hist(raw_data, bins=[40.5, 50.5, 55.5, 60.5, 70.5, 75.5], edgecolor="black")
```
