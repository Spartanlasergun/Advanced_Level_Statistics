```python
import matplotlib.pyplot as plt

midpoints = [182, 189, 194, 199, 204, 212]
frequencies = [12, 8, 8, 9, 7, 6]
raw_data = []
index = 0
for data in frequencies:
    count = data
    while count != 0:
        raw_data.append(midpoints[index])
        count = count - 1
    index = index + 1

y,edges,_ = plt.hist(raw_data, bins=[176.5, 186.5, 191.5, 196.5, 201.5, 206.5, 216.5], edgecolor="black")
midpoints = 0.5*(edges[1:]+edges[:-1]) 
plt.plot(midpoints, y)
    
```
