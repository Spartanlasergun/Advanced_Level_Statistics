# Question 1 (a)

This is our dataset

| x | f |
| --- | --- |
| 304 | 1 |
| 308 | 5 |
| 312 | 9 |
| 316 | 4 |
| 320 | 4 |
| 324 | 2 |

It is coded using the following equation:

$y = \frac{x - 312}{4}$


Hence:

$4y = x - 312$

$x = 4y + 312$

$\bar x = 4\bar y + 312$ 

and

$s_x = 4s_y$

Solving for $\bar y$:

$\bar y = \frac{\Sigma y}{n}$

$\bar y = \frac{(304) + (308 * 5) + (312 * 9) + (316 * 4) + (320 * 4) + (324 * 2)}{1 + 5 + 9 + 4 + 4 + 2}$

$\bar y = \frac{7844}{25}$

$\bar y = 313.76$

This implies that $\bar x = (4 * 313.76) + 312$

$\bar x = 1567.04$

Solving for $s_y$:

$s_y = \sqrt{\frac{\Sigma y^2}{n} - \bar y^2}$

... $s_y = 5.194$

Hence:

$s_x = 4 * 5.194$

$s_x = 20.776$
