```python
x = [2.3, 2.4, 2.5, 2.6, 2.7]
freq = [40, 60, 90, 50, 60]
raw_data = []
index = 0
for item in freq:
    count = item
    while count != 0:
        raw_data.append(x[index])
        count = count - 1
    index = index + 1
print(raw_data)
```

```python
import random
rand_sample = []
n = 3
while n != 0:
    get = random.choice(raw_data)
    rand_sample.append(get)
    n = n - 1
print(rand_sample)
```
