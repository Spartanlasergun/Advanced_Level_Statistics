```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

x = [39.5, 44.5, 49.5, 54.5, 59.5, 64.5]
y = [0, 8, 30, 64, 94, 120]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
x_list = x_new.tolist()
y_list = y_new.tolist()
upper_quartile_index = 120 * 0.75
lower_quartile_index = 120 * 0.25
for item in y_list:
    if item <= upper_quartile_index:
        upper_index = y_list.index(item)
upper_quartile = x_list[upper_index]

for item in y_list:
    if item <= lower_quartile_index:
        lower_index = y_list.index(item)
lower_quartile = x_list[lower_index]
intq_range = upper_quartile - lower_quartile
print("(a) The estimated interquartile range is = " + str(intq_range))
```

```python
index = 0
for item in x_list:
    if item <= 62:
        index = x_list.index(item)
corr_y = y_list[index]
total_people = 120
over_62 = int(total_people - corr_y)
percentage = (over_62 / 120) * 100
print("(b) The estmated number of people that spent over 62 minutes is = " + str(percentage) + "%")
```
