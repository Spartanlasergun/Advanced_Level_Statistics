```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# Define the data
x = [4.4, 4.8, 5.2, 5.6, 6.0, 6.4, 6.8, 7.2, 7.6, 8.0, 8.4]
y = [0, 1, 2, 5, 10, 19, 38, 43, 46, 49, 50]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Find the x value that corresponds to a given y value
x_list = x_new.tolist()
y_list = y_new.tolist()
for item in x_list:
    if item <= 7:
        index = x_list.index(item)
b = int(y_list[index])
percentage = (b / 50) * 100
print("The percentage of the samples that have a pH less than 7 is " + str(percentage) + "%") 
```

```python
# Find the y value that corresponds to a given x value
x_list = x_new.tolist()
y_list = y_new.tolist()
for item in y_list:
    if item <= 25:
        index = y_list.index(item)
center_pH = int(x_list[index])
print("50% of the sample had a pH greater than " + str(center_pH))
print("This is known in statistics as the median")
```
