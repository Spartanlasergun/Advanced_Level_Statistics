```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

x = [660, 675, 685, 695, 701]
y = [3, 7, 20, 17, 3]

f = interp1d(x, y, kind='linear')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Find the x value that corresponds to a given y value
x_list = x_new.tolist()
y_list = y_new.tolist()

median_index = int(len(x_list) / 2)
median = x_list[median_index]
print("(a) Based on a linear interpolation between the points, the median is = " + str(median))
```

```python
upper_quartile_index = int(len(x_list) * 0.75)
upper_quartile = x_list[upper_quartile_index]

lower_quartile_index = int(len(x_list) * 0.25)
lower_quartile = x_list[lower_quartile_index]

interquartile_range = upper_quartile - lower_quartile

print("(b) The interquartile range is = " + str(interquartile_range))
```
