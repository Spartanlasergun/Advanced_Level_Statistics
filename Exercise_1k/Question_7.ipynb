| Temperature(deg C) | Number of Weeks | Cumulative Frequency |
| --- | --- | --- |
| -5 to -1 | 8 | 8 |
| 0 to 4 | 12 | 20 |
| 5 to 9 | 17 | 37 |
| 10 to 14 | 31 | 68 |
| 15 to 19 | 23 | 91 |
| 20 to 24 | 9 | 100 |
| 25 to 29 | 4 | 104 |

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

x = [-3, 2, 7, 12, 17, 22, 27]
y = [8, 20, 37, 68, 91, 100, 104]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
x_list = x_new.tolist()
y_list = y_new.tolist()
median_index = int(len(y_list) / 2)
median_temp = x_list[median_index]
print("(b) The estimated median temperature is = " + str(median_temp))
```

```python
index_count = 0
for item in x_list:
    if item <= 17:
        index_count = index_count + 1
temp_on = y_list[index_count]
print("(c) The temperature was turned on for approximately " + str(temp_on) + " weeks")
```

```python
index_count = 0
for item in x_list:
    if item <= 21:
        index_count = index_count + 1
temp_warm = 104 - y_list[index_count]
percentage = (temp_warm / 104) * 100
print("(d) The temperature was extremely warm for approximately " + str(percentage) + "% of the time")
```
