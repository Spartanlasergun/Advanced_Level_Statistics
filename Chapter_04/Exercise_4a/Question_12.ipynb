$$let \ \ P(X = 9) = a, \ \ hence \ \ P(X = 8) = 2a$$

$$0.1 + 0.2 + 0.4 + a + 2a = 1$$

$$0.7 + 3a = 1$$

$$3a = 0.3$$

$$a = 0.1$$

Hence, 

$$P(X = 8) = 0.2, \ \ P(X = 9) = 0.1$$

```python
import matplotlib.pyplot as plt

x = [5, 6, 7, 8, 9]
y = [0.1, 0.2, 0.4, 0.2, 0.1]

plt.bar(x, y, width=0.1)
plt.title("Symmetrical Distribution")
plt.show()
```
