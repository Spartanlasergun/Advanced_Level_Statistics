```python
import math

def bino(n, r, p, q):
    funk = ((math.factorial(n)) / (math.factorial(r) * math.factorial((n-r)))) * pow(q, (n-r)) * pow(p, r)
    return funk
```

```python
n = 6
p = 0.5
q = 1 - p

dist = []
r = -1
while r!= 6:
    r = r + 1
    calc = bino(n, r, p, q)
    dist.append(calc)

x = [0, 1, 2, 3, 4, 5, 6]
```

```python
import matplotlib.pyplot as plt

plt.bar(x, dist, width=0.1)
plt.show()
```
