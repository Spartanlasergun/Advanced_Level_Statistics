```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# Define the data
x = [4.5, 13.5, 22.5, 31.5, 40.5]
y = [6, 16, 37, 57, 60]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

# Find the x value that corresponds to a given y value
x_list = x_new.tolist()
y_list = y_new.tolist()
for item in x_list:
    if item <= 30:
        index = x_list.index(item)
b = int(y_list[index])
print(b)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```
