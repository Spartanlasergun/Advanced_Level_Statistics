```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# Define the data
x = [515, 540, 560, 585, 625]
y = [15, 39, 72, 93, 100]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

# Find the x value that corresponds to a given y value
x_list = x_new.tolist()
y_list = y_new.tolist()
for item in x_list:
    if item <= 540:
        index = x_list.index(item)
b540 = int(y_list[index])
print("540 = " + str(b540))

# Find the x value that corresponds to a given y value
x_list = x_new.tolist()
y_list = y_new.tolist()
for item in x_list:
    if item <= 580:
        index = x_list.index(item)
b580 = int(y_list[index])
print("580 = " + str(b580))

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```
