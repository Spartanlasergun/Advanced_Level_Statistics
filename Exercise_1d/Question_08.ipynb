```python
import matplotlib.pyplot as plt

data = [14.5, 22, 27, 30, 32.5, 37, 49.5]
freq = [20, 20, 15, 14, 16, 10, 10]
raw_data = []
index = 0
for item in freq:
    count = item
    while count != 0:
        raw_data.append(data[index])
        count = count - 1
    index = index + 1
    
plt.hist(raw_data, bins=[9.5, 19.5, 24.5, 29.5, 30.5, 34.5, 39.5, 59.5], edgecolor="black")
mean = (sum(raw_data)) / (len(raw_data))
print("The mean is = " + str(mean))
print("You would represent this data using a histogram because the intervals for measurement are genuinely non-uniform")
```
