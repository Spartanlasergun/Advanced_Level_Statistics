```python
import matplotlib.pyplot as plt

data = [7.5, 22.5, 40, 60, 85]
freq = [54, 78, 120, 88, 60]
raw_data = []
index = 0
for item in freq:
    count = item
    while count != 0:
        raw_data.append(data[index])
        count = count - 1
    index = index + 1
    
plt.hist(raw_data, bins=[0, 15, 30, 50, 70, 100], edgecolor="black")
mean = (sum(raw_data)) / (len(raw_data))
print("The mean is = " + str(mean))
```
