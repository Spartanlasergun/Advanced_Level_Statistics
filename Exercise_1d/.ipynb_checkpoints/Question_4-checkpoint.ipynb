```python
import matplotlib.pyplot as plt

data = [2.5, 7.5, 12.5, 17.5, 30]
freq = [12, 38, 42, 20, 8]
raw_data = []
index = 0
for item in freq:
    count = item
    while count != 0:
        raw_data.append(data[index])
        count = count - 1
    index = index + 1
    
plt.hist(raw_data, bins=[0, 5, 10, 15, 20, 40], edgecolor="black")
mean = (sum(raw_data)) / (len(raw_data))
print("The mean is = " + str(mean))
```
