```python
import matplotlib.pyplot as plt

data = [7, 11, 14, 17, 23.5]
freq = [14, 18, 15, 4, 8]
raw_data = []
index = 0
for item in freq:
    count = item
    while count != 0:
        raw_data.append(data[index])
        count = count - 1
    index = index + 1
    
plt.hist(raw_data, bins=[4.5, 9.5, 12.5, 15.5, 18.5, 28.5], edgecolor="black")
mean = (sum(raw_data)) / (len(raw_data))
mean = round(mean, 1)
print("The mean is = " + str(mean))
```
