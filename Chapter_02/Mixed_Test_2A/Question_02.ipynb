```python
Et = 44
Et_2 = 180.4
Ety = 824.5
Ey = 205

Sxy = Ety - ((Et * Ey) / 11)
Sxx = Et_2 - ((pow(Et, 2)) / 11)
b = Sxy / Sxx
c = (Ey / 11) - (b * (Et / 11))
print("regression function is given by y = " + str(b) + "x + " + str(c))
```

the y intercept indicates the initial temperature of the bottle of milk

```python
at45 = (b * 4.5) + c
print("Estimated temp at 4.5 mins = " + str(at45))
```

```python
at20 = (b * 20) + c
print("Estimated temp at 20 mins = " + str(at20))
```

The estimate at 4.5 miniutes is more reliable because it lies within the dataset
