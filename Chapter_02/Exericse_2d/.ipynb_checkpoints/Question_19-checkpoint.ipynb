```python
import matplotlib.pyplot as plt

x = [18, 36, 45, 22, 69, 72, 13, 33, 59, 79, 10, 53]
y = [37, 54, 63, 42, 84, 91, 33, 49, 79, 98, 32, 71]

plt.scatter(x, y)
plt.grid(True)
plt.show()
```

The regression line is a suitable model, because the scatterplot indicates that the spread of the data is very linear.

```python
import statistics

var = statistics.variance(x)
covar = statistics.covariance(x, y)
gradient = covar / var
mean_x = statistics.mean(x)
mean_y = statistics.mean(y)
y_intercept = mean_y - (gradient * mean_x)
print("regression function is given by y = " + str(gradient) + "x + " + str(y_intercept))
```

```python
print("or, " + "(y - " + str(mean_y) + ") = "+ str(gradient) + "(x - " + str(mean_x) + ")")
```
