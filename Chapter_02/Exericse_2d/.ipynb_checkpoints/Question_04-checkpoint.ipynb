```python
import matplotlib.pyplot as plt

x = [40, 42, 44, 46, 48, 50, 52, 54]
y = [12.8, 11.6, 11.3, 10.3, 10.7, 9.1, 8.9, 9.2]

plt.scatter(x, y)
plt.grid(True)
plt.show()
```

```python
import statistics

var = statistics.variance(x)
covar = statistics.covariance(x, y)
gradient = covar / var
mean_x = statistics.mean(x)
mean_y = statistics.mean(y)
y_intercept = mean_y - (mean_x * gradient)
print("regression function is given by y = " + str(gradient) + "x + " + str(y_intercept))
```

```python
rx = list(range(40, 54, 1))
ry = []
for item in rx:
    fx = (gradient * item) + y_intercept
    ry.append(fx)

plt.scatter(x, y)
plt.plot(rx, ry)
plt.grid(True)
plt.show()
```

```python
at58 = (gradient * 58) + y_intercept
print("Price at 58 pounds = " + str(at58))
```

The estimate at 58 pounds is not reliable as it lies outside the scope of the dataset
