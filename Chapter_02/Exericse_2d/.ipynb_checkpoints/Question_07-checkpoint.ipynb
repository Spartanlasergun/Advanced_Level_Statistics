```python
from scipy.stats import pearsonr

x = [14, 13, 17, 16, 17, 18, 19, 32, 18, 15, 15, 19]
y = [7.2, 7.3, 6.9, 7.3, 7.5, 7.6, 6.8, 3.7, 6.5, 7.9, 6.8, 7.1]

pearr = pearsonr(x, y)[0]
print("Pearson's r = " + str(pearr))
```

A strong negative correlation indicates that the more items completed by a seamstress, the lower the quality of the work done.

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.grid(True)
plt.show()
```
