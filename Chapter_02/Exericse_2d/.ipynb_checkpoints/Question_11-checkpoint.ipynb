```python
import matplotlib.pyplot as plt

n = [11, 21, 28, 37, 75]
p = [11, 26, 34, 41, 88]

plt.scatter(n, p)
plt.grid(True)
plt.show()
```

```python
import statistics

var = statistics.variance(n)
covar = statistics.covariance(n, p)
gradient = covar / var
mean_n = statistics.mean(n)
mean_p = statistics.mean(p)
y_intercept = mean_p - (gradient * mean_n)
print("regression function is given by y = " + str(gradient) + "x + " + str(y_intercept))
```

```python
at15 = (gradient * 15) + y_intercept
print("estimated price for set 2 = " + str(at15))
```

```python
rx = list(range(11, 75, 1))
ry = []
for item in rx:
    fx = (gradient * item) + y_intercept
    ry.append(fx)
plt.scatter(n, p)
plt.plot(rx, ry)
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr

pearr = pearsonr(n, p)[0]
print("Pearson's r = " + str(pearr))
```

The scatter diagram and pearson's coefficient indicate that as number of pieces increase, price increases.
