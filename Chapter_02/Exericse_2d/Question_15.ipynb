```python
import matplotlib.pyplot as plt

x = [1.32, 1.54, 1.45, 1.06, 2.13, 1.00, 0.90, 2.00, 2.70, 0.75]
y = [55, 48, 55, 53, 74, 44, 49, 68, 78, 51]

plt.scatter(x, y)
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr

pearr = pearsonr(x, y)[0]
print("Pearson's r = " + str(pearr))
```

The pearson's coefficient indicates a strong positive correlation between body mass and creatine output.
