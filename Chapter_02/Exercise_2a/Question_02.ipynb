```python
import matplotlib.pyplot as plt

x = [110, 120, 130, 140, 150, 160, 170]
y = [2.1, 4.3, 3.1, 3.4, 2.9, 5.5, 3.3]

plt.scatter(x, y)
plt.plot()
```

Based on the spread of the data, I would say that a linear regression model is appropriate

```python
import statistics

var = statistics.variance(x)
covar = statistics.covariance(x, y)
mean_x = statistics.mean(x)
mean_y = statistics.mean(y)
gradient = covar / var
y_intercept = mean_y - (gradient * mean_x)

reg_x = list(range(100, 180, 1))
reg_y = []
for item in reg_x:
    func = (gradient * item) + y_intercept
    reg_y.append(func)

print("The regression function is given by y = " + str(gradient) + "x + " + str(y_intercept))
    
plt.scatter(x, y)
plt.plot(reg_x, reg_y)
plt.show()
```
