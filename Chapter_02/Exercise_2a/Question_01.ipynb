```python
import statistics

x_data_1 = [3, 7, 9, 11, 14, 14, 15, 21, 22, 23, 26]
y_data_1 = [5, 12, 5, 12, 10, 17, 23, 16, 10, 20, 25]

var = statistics.variance(x_data_1)
covar = statistics.covariance(x_data_1, y_data_1)
x_mean = statistics.mean(x_data_1)
y_mean = statistics.mean(y_data_1)
gradient = covar / var
y_intercept = y_mean - (x_mean * gradient)
print("regression equation for dataset(y vs x) one is given by y = " + str(gradient) + "x + " + str(y_intercept))

var = statistics.variance(y_data_1)
covar = statistics.covariance(y_data_1, x_data_1)
x_mean = statistics.mean(y_data_1)
y_mean = statistics.mean(x_data_1)
gradient_1 = covar / var
y_intercept_1 = y_mean - (x_mean * gradient)
print("regression equation for dataset(x vs y) one is given by y = " + str(gradient_1) + "x + " + str(y_intercept_1))
```

```python
import matplotlib.pyplot as plt

x = list(range(min(x_data_1), max(x_data_1), 1))
y = []
for item in x:
    func = (gradient * item) + y_intercept
    y.append(func)

y_2 = []
for item in x:
    func = (gradient_1 * item) + y_intercept_1
    y_2.append(func)
         
plt.scatter(x_data_1, y_data_1)
plt.plot(x, y)
plt.plot(y_2, x)
plt.show()
```

The graph shows a good positive correlation

```python
x_data_2 = [1, 5, 5, 5, 6, 7.5, 7.5, 7.5, 10, 11, 12.5, 14, 14.5]
y_data_2 = [85, 82, 85, 89, 78, 66, 77, 81, 70, 74, 65, 69, 63]

var_2 = statistics.variance(x_data_2)
covar_2 = statistics.covariance(x_data_2, y_data_2)
x_mean_2 = statistics.mean(x_data_2)
y_mean_2 = statistics.mean(y_data_2)
gradient_2 = covar_2 / var_2
y_intercept_2 = y_mean_2 - (x_mean_2 * gradient_2)
print("regression equation for dataset two(y vs x) is given by y = " + str(gradient_2) + "x + " + str(y_intercept_2))

var_3 = statistics.variance(y_data_2)
covar_3 = statistics.covariance(y_data_2, x_data_2)
x_mean_3 = statistics.mean(y_data_2)
y_mean_3 = statistics.mean(x_data_2)
gradient_3 = covar_3 / var_3
y_intercept_3 = y_mean_3 - (x_mean_3 * gradient_3)
print("regression equation for dataset two(x vs y) is given by y = " + str(gradient_3) + "x + " + str(y_intercept_3))
```

```python
import matplotlib.pyplot as plt

x = list(range(1, 100, 1))
y = []
for item in x:
    func = (gradient_2 * item) + y_intercept_2
    y.append(func)

y_2 = []
for item in x:
    func = (gradient_3 * item) + y_intercept_3
    y_2.append(func)
         
plt.scatter(x_data_2, y_data_2)
plt.plot(x, y)
plt.plot(y_2, x)
plt.show()
```
