```python
import matplotlib.pyplot as plt

x = [14, 29, 55, 74, 11, 23, 47, 69, 18, 36, 61, 79]
y = [35, 50, 73, 93, 31, 42, 65, 86, 38, 54, 81, 96]

plt.scatter(x, y)
plt.plot()
```

```python
import statistics

var = statistics.variance(x)
covar = statistics.covariance(x, y)
mean_x = statistics.mean(x)
mean_y = statistics.mean(y)
gradient = covar / var
y_intercept = mean_y - (gradient * mean_x)
print("The regression function is given by y = " + str(gradient) + "x + " + str(y_intercept))
```

```python
reg_x = list(range(0, 90, 1))
reg_y = []
for item in reg_x:
    func = (gradient * item) + y_intercept
    reg_y.append(func)

plt.scatter(x, y)
plt.plot(reg_x, reg_y)
plt.show()
```
