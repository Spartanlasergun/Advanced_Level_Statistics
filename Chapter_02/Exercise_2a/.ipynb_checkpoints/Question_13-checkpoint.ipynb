```python
import matplotlib.pyplot as plt

x = [2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0]
y = [127.6, 130.2, 132.7, 133.6, 133.9, 133.8, 133.3, 131.9]

plt.scatter(x, y)
plt.show()
```

```python
import statistics

var = statistics.variance(x)
covar = statistics.covariance(x, y)
mean_x = statistics.mean(x)
mean_y = statistics.mean(y)
gradient = covar / var
y_intercept = mean_y - (gradient * mean_x)
print("The regression function is given by y = " + str(gradient) + "x + " + str(y_intercept))
```

```python
reg_x = list(range(0, 8, 1))
reg_y = []
for item in reg_x:
    func = (gradient * item) + y_intercept
    reg_y.append(func)

plt.scatter(x, y)
plt.plot(reg_x, reg_y)
plt.show()
```

```python
a = [70, 75, 80, 85, 90]
b = [127.6, 128.7, 130.4, 131.2, 133.6]

reg_a = list(range(65, 95, 1))
reg_b = []
for item in reg_a:
    func = (0.29 * item) + 107.1
    reg_b.append(func)

plt.scatter(a, b)
plt.plot(reg_a, reg_b)
plt.show()
```
