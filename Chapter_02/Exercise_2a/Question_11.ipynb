```python
import matplotlib.pyplot as plt

x = [10, 55, 46, 27, 17, 12, 85, 64, 36, 40, 30, 37]
y = [5750, 17300, 14750, 8200, 6350, 6150, 18800, 14850, 9990, 11000, 9150, 10400]

plt.scatter(x, y)
plt.show()
```

```python
import statistics
var = statistics.variance(x)
covar = statistics.covariance(x, y)
mean_x = statistics.mean(x)
mean_y = statistics.mean(y)
b = covar / var
a = mean_y - (b * mean_x)
print("a = " + str(a))
print("b = " + str(b))
```

```python
reg_x = list(range(0, 100, 1))
reg_y = []
for item in reg_x:
    func = (b * item) + a
    reg_y.append(func)

plt.scatter(x, y)
plt.plot(reg_x, reg_y)
plt.show()
```
