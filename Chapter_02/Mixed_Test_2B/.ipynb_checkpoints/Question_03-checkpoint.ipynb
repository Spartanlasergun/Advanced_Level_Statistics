```python
import matplotlib.pyplot as plt

x = [30, 50, 70, 90, 110, 130]
y = [25, 50, 85, 155, 235, 350]

plt.scatter(x, y)
plt.grid(True)
plt.show()
```

```python
import statistics

var = statistics.variance(x)
covar = statistics.covariance(x, y)
gradient = covar / var
mean_x = statistics.mean(x)
mean_y = statistics.mean(y)
y_intercept = mean_y - (gradient * mean_x)
print("regression function is given by y = " + str(gradient) + " x + " + str(y_intercept))
```

```python
rx = list(range(min(x), max(x), 1))
ry = []
for item in rx:
    fx = (gradient * item) + y_intercept
    ry.append(fx)

plt.scatter(x, y)
plt.plot(rx, ry)
plt.grid(True)
plt.show()
```

```python
x100 = (gradient * 100) + y_intercept
print("at x = 100, y = " + str(x100))
```

```python
x150 = (gradient * 150) + y_intercept
print("at x = 150, y = " + str(x150))
```

The prediction at x = 100 is more accurate that the prediction at x = 150, because the former lies within the range of the dataset , while the latter is beyond the range of the dataset. 

The regression line may be a farily reliable estimate based on the scatterplot, but the relationship may be more accurately modeled in a non-linear context.
