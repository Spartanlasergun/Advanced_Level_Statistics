```python
import matplotlib.pyplot as plt

x = [10, 11, 2, 7, 5, 6, 12]
y = [17, 21, 12, 13, 18, 16, 15]

plt.scatter(x, y)
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr

pearr = pearsonr(x, y)[0]

print("Pearson's r = " + str(pearr))
```
