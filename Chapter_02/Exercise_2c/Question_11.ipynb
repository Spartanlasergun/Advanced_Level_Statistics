```python
from scipy.stats import spearmanr

x = [1.26, 1.25, 0.65, 2.10, 2.45, 2.17, 2.84, 1.74, 2.57, 1.65, 1.47, 1.94]
y = [1.1, 2.7, 4.5, 5.1, 5.5, 7.6, 5.2, 5.7, 4.8, 2.9, 2.8, 1.8]

spearr = spearmanr(x, y)[0]
print("Spearman's r = " + str(spearr))
```

The rank order would remain the same for hours and minutes. Hence, spearman's r for values in miniutes is still 0.601
