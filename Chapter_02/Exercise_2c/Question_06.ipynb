```python
import matplotlib.pyplot as plt

p = [55, 11, 68, 38, 46, 43, 21, 25]
d = [0.7, 3.8, 1.7, 2.6, 1.5, 2.6, 3.4, 1.9]

plt.scatter(p, d)
plt.grid(True)
plt.show()
```

```python
import statistics

mean_p = statistics.mean(p)
mean_d = statistics.mean(d)

plt.scatter(p, d)
plt.scatter(mean_p, mean_d)
plt.text(mean_p, mean_d + 0.1, '(Mean)', ha='center')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import spearmanr

spearr = spearmanr(p, d)[0]
print("Spearman's r = " + str(spearr))
```

The spearman's coefficient shows strong negative correlation, meaning that as you go futher from the center of the town, the population density decreases.

C fits least well, because it has the highest population density and is roughly halfway away from the center of the town.
