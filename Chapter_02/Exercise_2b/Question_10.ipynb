```python
import matplotlib.pyplot as plt

x = [27, 30, 37, 38, 32, 36, 32, 32, 38, 42, 36, 44, 33, 38]
y = [118, 136, 156, 150, 140, 155, 157, 114, 144, 159, 149, 170, 131, 160]

plt.scatter(x, y)
plt.show()
```

```python
import statistics

var = statistics.variance(x)
covar = statistics.covariance(x, y)
mean_x = statistics.mean(x)
mean_y = statistics.mean(y)
gradient = covar / var
y_intercept = mean_y - (mean_x * gradient)
print("The regression equation is given by y = " + str(gradient) + "x +" + str(y_intercept))
```

```python
Rx = list(range(min(x), max(x), 1))
Ry = []
for item in Rx:
    fx = (gradient * item) + y_intercept
    Ry.append(fx)

plt.scatter(x, y)
plt.plot(Rx, Ry)
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr

r = pearsonr(x, y)[0]
print("r = " + str(r))
```
