$$1.25 - \frac{1.25}{x} = 0.5$$

$$\frac{1.25}{x} = 0.75$$

$$0.75x = 1.25$$

$$x = 1.66667$$

$$median = 1.66667 miles$$

```python
from scipy.integrate import quad

# Define the function to integrate
def integrand(x):
    return (1.25 / (x**2))

# Set the integration limits
lower_limit = 1.16667
upper_limit = 2.16667

# Perform the integration
result, error = quad(integrand, lower_limit, upper_limit)

# Print the result
print("P(1.16667 <= x <= 2.16667) = " + str(result))
```
