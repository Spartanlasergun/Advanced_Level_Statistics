```python
from scipy.integrate import quad

# Define the function to integrate
def integrand(x):
    return 0.25 * x**4

# Set the integration limits
lower_limit = 0
upper_limit = 2

# Perform the integration
EX, error = quad(integrand, lower_limit, upper_limit)

# Print the result
print("E(X) = " + str(EX))
```

```python
def integrand(x):
    return 0.25 * x**5

# Set the integration limits
lower_limit = 0
upper_limit = 2

# Perform the integration
EX_2, error = quad(integrand, lower_limit, upper_limit)
Var = EX_2 - (EX**2)
import math
std = math.sqrt(Var)
print("std = " + str(std))
```
