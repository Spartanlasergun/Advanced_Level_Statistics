a)

$$f(x) = \frac{k}{x}$$

$$\int_{1}^{9} \frac{k}{x} \ dx = 1$$

$$k [\ln (x)]_{1}^{9} = 1$$

$$k (\ln(9) - \ln(1)) = 1$$

$$k = 0.455$$

$$F(x) = 0.455\ln(x)$$

for the median,

$$0.455\ln(x) = 0.5$$

$$\ln(x) = 1.0989$$

$$x = e^{1.0989}$$

$$x = 3$$

$$hence, \ \ \ median = 3$$

b)

$$f(x) = \frac{0.455}{x}$$

$$xf(x) = 0.455$$

$$E(X) = \int_{1}^{9} 0.455 \ dx$$

$$E(X) = [0.455x]_{1}^{9}$$

$$E(X) = 3.64$$

$$x^2f(x) = 0.455x$$

$$E(X^2) = \int_{1}^{9} 0.455x \ dx$$

$$E(X^2) = [0.2275x^2]_{1}^{9}$$

$$E(X^2) = 18.4275 - 0.2275$$

$$E(X^2) = 18.2$$

$$Var(X) = E(X^2) - (E(X))^2$$

$$Var(X) = 18.2 - (3.64^2)$$

$$Var(X) = 4.95$$

c)

$$f(x) = \frac{0.455}{x}$$

$$F(x) = 0.455 \ln(x)$$

```python
import math
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 9.1, 0.1)
y = []
for item in x:
    fx = 0.455 * math.log(item)
    y.append(fx)

plt.plot(x, y)
plt.show()
```
