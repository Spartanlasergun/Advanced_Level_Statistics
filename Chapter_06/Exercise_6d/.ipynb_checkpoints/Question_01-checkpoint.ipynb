$$f(x) = \frac{3x^2}{8}$$

$$F(X) = \int_{0}^{t} \frac{3x^2}{8} \ dx$$

$$F(X) = \frac{t^3}{8}$$

-----

$$\frac{t^3}{8} = 0.5$$

$$t^3 = 4$$

$$t = 1.59$$

hence, median = 1.59

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2.1, 0.1)
y = []
for item in x:
    fx = pow(item, 3) / 8
    y.append(fx)

plt.plot(x, y)
plt.show()
```
