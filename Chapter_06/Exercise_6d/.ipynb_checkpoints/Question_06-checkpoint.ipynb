a)

$$f(x) = \frac{1}{3}$$

$$xf(x) = \frac{x}{3}$$

$$E(X) = \int_{0}^{3} \frac{x}{3} \ dx$$

$$E(X) = [\frac{x^2}{6}]_{0}^{3}$$

$$E(X) = 1.5$$

-----
b)

$$x^2f(x) = \frac{x^2}{3}$$

$$E(X^2) = \int_{0}^{3} \frac{x^2}{3} \ dx$$

$$E(X^2) = [\frac{x^3}{9}]_{0}^{3}$$

$$E(X^2) = 3$$

$$Var(X) = E(X^2) - (E(X))^2$$

$$Var(X) = 3 - (1.5)^2$$

$$Var(X) = 0.75$$

c)

$$f(x) = \frac{1}{3}$$

$$F(x) = \frac{x}{3}$$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 3.1, 0.1)
y = []
for item in x:
    fx = item / 3
    y.append(fx)

plt.plot(x, y)
plt.show()
```
