$$f(x) = ax - bx^2$$

$$\int_{0}^{2} f(x) \ dx = 1$$

$$[\frac{ax^2}{2} - \frac{bx^3}{3}]_{0}^{2} = 1$$

$$(2a - \frac{8b}{3}) = 1$$

$$6a - 8b = 3, \ \ \ \ ---equation \ one$$

$$E(X) = 1$$

$$xf(x) = ax^2 - bx^3$$

$$\int_{0}^{2} xf(x) \ dx = 1$$

$$[\frac{ax^3}{3} - \frac{bx^4}{4}]_{0}^{2} = 1$$

$$\frac{8a}{3} - 12b = 3$$

$$8a - 12b = 3, \ \ \ \ ---equation \ two$$

solving for b,

$$6a = 3 + 8b$$

$$a = \frac{3 + 8b}{6}$$

hence,

$$\frac{24 + 64b}{6} - 12b = 3$$

$$24 + 64b - 72b = 18$$

$$8b = 6$$

$$b = 0.75$$

solving for a,

$$a = \frac{3 + (8 * 0.75)}{6}$$

$$a = 1.5$$
