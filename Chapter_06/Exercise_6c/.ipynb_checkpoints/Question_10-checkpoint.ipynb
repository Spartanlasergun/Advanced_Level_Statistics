```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 3.1, 0.1)
y = []
for item in x:
    fx = (1/27) * pow(item, 2)
    y.append(fx)

y_2 = [(1/3), (1/3)]
x_2 = [3, 5]
plt.plot(x, y)
plt.plot(x_2, y_2)
plt.show()
```

b)

$$xf(x) = \frac{x^3}{27}$$

$$xf(x) = \frac{x}{3}$$

$$E(X) = \int_{0}^{3} \frac{x^3}{27} \ dx + \int_{3}^{5} \frac{x}{3} \ dx$$

$$E(X) = [\frac{x^4}{108}]_{0}^{3} + [\frac{x^2}{6}]_{3}^{5}$$

$$E(X) = 0.75 + (4.166667 - 1.5)$$

$$E(X) = 3.41666$$

-----
c)

$$x^2f(x) = \frac{x^4}{27}$$

$$x^2{fx} = \frac{x^2}{3}$$

$$E(X^2) = \int_{0}^{3} \frac{x^4}{27} \ dx + \int_{3}^{5} \frac{x^2}{3} \ dx$$

$$E(X^2) = [\frac{x^5}{135}]_{0}^{3} + [\frac{x^3}{9}]_{3}^{5}$$

$$E(X^2) = 1.8 + (13.88888 - 3)$$

$$E(X^2) = 12.68888$$

-----
d)

$$Var(X) = E(X^2) - (E(X))^2$$

$$Var(X) = 12.68888 - (3.4166^2)$$

$$Var(X) = 1.016$$

$$\sigma = \sqrt(1.016)$$

$$\sigma = 1.0078$$
