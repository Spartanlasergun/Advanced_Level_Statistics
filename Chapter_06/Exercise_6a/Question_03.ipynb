$$\int_{1}^{3} k(4 - x) = 1$$

$$\int_{1}^{3} 4k - kx = 1$$

$$[4kx - \frac{kx^2}{2}]_{1}^{3} = 1$$

$$(12k - \frac{9k}{2}) - (4k - \frac{k}{2}) = 1$$

$$7.5k - 3.5k = 1$$

$$4k = 1$$

$$k = 0.25$$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 3.1, 0.1)
y = []
for item in x:
    fx = 0.25 * (4 - item)
    y.append(fx)

plt.plot(x, y, linestyle="--")
plt.show()
```
