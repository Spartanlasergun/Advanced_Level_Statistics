b)

$$E(X) = \int_{-2}^{3} 0.2x \ dx$$

$$E(X) = [\frac{0.1x^2}]_{-2}^{3}$$

$$E(X) = (0.9) - (0.4)$$

$$E(X) = 0.5$$

$$E(X^2) = \int_{-2}^{3} 0.2x^2 \ dx$$

$$E(X^2) = [\frac{0.2x^3}{3}]_{-2}^{3}$$

$$E(X^2) = (1.8) - (-0.5333)$$

$$E(X^2) = 2.3333$$

$$Var(X) = E(X^2) - (E(X))^2$$

$$Var(X) = 2.3333 - (0.5^2)$$

$$Var(X) = 2.08333$$

$$\sigma = \sqrt(2.083333)$$

$$\sigma = 1.44$$
