```python
from scipy.integrate import quad

def integrand(x):
    return (0.5 * x)

def integrand_2(x):
    return (2 / 3) - (x / 6)

# Set the integration limits
lower_limit = 0
upper_limit = 1

# Perform the integration
result_1, error = quad(integrand, lower_limit, upper_limit)
E1 = result_1 * 200

result_2, error = quad(integrand_2, 1, 2)
E2 = result_2 * 250

result_3, error = quad(integrand_2, 2, 3)
E3 = result_3 * 350

result_4, error = quad(integrand_2, 3, 4)
E4 = result_4 * 500

EX = E1 + E2 + E3 + E4

print("E(X) = " + str(EX))
```
