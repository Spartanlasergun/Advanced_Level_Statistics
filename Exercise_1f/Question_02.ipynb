```python
import matplotlib.pyplot as plt
import statistics

frequency = [10, 16, 40, 26, 8]
midpoints = [225, 275, 337.5, 387.5, 450]
raw_data = []
index = 0
for item in frequency:
    count = item
    while count != 0:
        raw_data.append(midpoints[index])
        count = count - 1
    index = index + 1
plt.hist(raw_data, bins=[200, 250, 300, 375, 400, 500], edgecolor="black")

mean = (sum(raw_data)) / (len(raw_data))
print("The mean wage is = $" + str(mean))

std = statistics.pstdev(raw_data)
print("The Standard Deviation for wages = $" + str(std))
```
