# Question Fifteen

| Age (years) | Number of Applicants | Cum. Frequency |
| --- | --- | --- |
| 20- | 14 | 14 |
| 35- | 12 | 26 |
| 40- | 7 | 33 |
| 45- | 8 | 41 |
| 50- | 9 | 50 |
| 60- | 0 | 50 |

```python
import matplotlib.pyplot as plt

midpoints = [27.5, 37.5, 42.5, 47.5, 55, 65]
freq = [14, 12, 7, 8, 9, 0]
raw_data = []
index = 0

for item in freq:
    count = item
    while count != 0:
        raw_data.append(midpoints[index])
        count = count - 1
    index = index + 1

plt.hist(raw_data, bins=[20, 35, 40, 45, 50, 60], edgecolor='black')
plt.show()
```

```python
import statistics

mean = statistics.mean(raw_data)
print("mean = " + str(mean))
```

```python
from scipy.interpolate import interp1d
import numpy as np

intervals = [20, 35, 40, 45, 50, 60]
cum_freq = [14, 26, 33, 41, 50, 50]

x = np.linspace(20, 60, 100)
function = interp1d(intervals, cum_freq, kind='quadratic')
y = function(x)

plt.plot(x, y)
plt.show()
```

```python
x_list = x.tolist()
y_list = y.tolist()

index_count = 0
for item in x_list:
    if item <= 50:
        index_count = index_count + 1
median_50 = int(index_count / 2)
median = x_list[median_50]
print("The median age of applicant less that 50 years old is = " + str(median) + " years")
```
