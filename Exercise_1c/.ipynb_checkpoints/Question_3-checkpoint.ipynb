```python
import matplotlib.pyplot as plt

labels = ["A", "B", "C", "D", "E"]
data = [55, 130, 20, 35, 60]

plt.pie(data, labels=labels)
```

```python
import math
total = sum(data)
r_1 = 5
F_1 = total
F_2 = (total * 0.20) + total
r_2 = ((math.sqrt(F_2)) / (math.sqrt(F_1))) * r_1
print("The radius of the pie chart for the year 1989 = " + str(r_2))
```

```python
r_1 = r_2
F_1 = F_2
F_2 = (F_1 * 0.20) + F_1
r_2 = ((math.sqrt(F_2)) / (math.sqrt(F_1))) * r_1
print("The radius of the pie chart for the year 1990 = " + str(r_2))
```

```python
angle = (60 / F_2) * 360
print(angle)
```
