```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

x = [400, 410, 420, 430, 440, 450, 460]
y = [0, 4, 7, 13, 23, 27, 29]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
x_list = x_new.tolist()
y_list = y_new.tolist()
total = 29
median_val = total / 2

for item in y_list:
    if item <= median_val:
        index = y_list.index(item)
median = x_list[index]
print("The estimated median value is = " + str(median))
```

```python
index10 = total * 0.1

for item in y_list:
    if item <= index10:
        indexten = y_list.index(item)
percentile10 = x_list[indexten]
print("The estimated 10th percentile is = " + str(percentile10))
```

```python
index90 = total * 0.9

for item in y_list:
    if item <= index90:
        indexninety = y_list.index(item)
percentile90 = x_list[indexninety]
print("The estimated 90th percentile is = " + str(percentile90))
```
