| --- | --- | --- |

| 70-74 | 8 | 8 |
| 75-79 | 20 | 28 |
| 80-84 | 25 | 53 |
| 85-89 | 31 | 84 |
| 90-94 | 10 | 94 |
| 95-99 | 6 | 100 |

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

x = [69.5, 74.5, 79.5, 84.5, 89.5, 94.5, 99.5]
y = [0, 8, 28, 53, 84, 94, 100]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
x_list = x_new.tolist()
y_list = y_new.tolist()
for item in y_list:
    if item <= 25:
        l_index = y_list.index(item)
lower_quartile = x_list[l_index]

for item in y_list:
    if item <= 75:
        u_index = y_list.index(item)
upper_quartile = x_list[u_index]

intq_range = upper_quartile - lower_quartile
print("The interquartile range is = " + str(intq_range) + " seconds")
```

```python
for item in x_list:
    if item <= 78:
        index = x_list.index(item)
qualified = int(y_list[index])
print("The number of people that qualified are = " + str(qualified))
```

```python
for item in y_list:
    if item <= 38.5:
        index = y_list.index(item)
median = x_list[index]
print("The median time for those who qualified is = " + str(median) + " seconds")
```
