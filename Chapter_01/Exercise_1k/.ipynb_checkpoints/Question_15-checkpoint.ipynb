```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

x = [0, 25, 50, 64, 74]
y = [0, 20, 40, 60, 80]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
x_list = x_new.tolist()
y_list = y_new.tolist()

for item in y_list:
    if item <= 50:
        index = y_list.index(item)
median = x_list[index]
print("The estimated median is = " + str(median) + "minutes")
```

```python
for item in y_list:
    if item <= 75:
        index = y_list.index(item)
upper_quartile = x_list[index]
print("The estimated upper quartile is = " + str(upper_quartile) + "minutes")
```

```python
for item in x_list:
    if item <= 40:
        index = x_list.index(item)
less_than_40 = y_list[index]
print("The estimated percent of people who finsihed in less than 40 minutes is = " + str(less_than_40) + "%")
```
