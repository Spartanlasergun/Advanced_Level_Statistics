```python
import matplotlib.pyplot as plt

freq = [2, 5, 9, 8, 1, 0]
midpoints = [10, 25, 35, 45, 55, 65]
raw_data = []
index = 0
for item in freq:
    count = item
    while count != 0:
        raw_data.append(midpoints[index])
        count = count - 1
    index = index + 1
plt.hist(raw_data, bins=midpoints, density=True, edgecolor="black")
```

## Cumulative Frequency Table

| Thickness (mm) | Number of strata | Cumulative Frequency |
| --- | --- | --- |
| 0- | 2 | 2 |
| 20- | 5 | 7 |
| 30- | 9 | 16 |
| 40- | 8 | 24 |
| 50- | 1 | 25 |
| 60- | 0 | 25 |

```python
import numpy as np
from scipy.interpolate import interp1d

x = [0, 20, 30, 40, 50, 60]
y = [0, 2, 7, 16, 24, 25]

f = interp1d(x, y, kind='linear')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
x_list = x_new.tolist()
y_list = y_new.tolist()
for item in y_list:
    if item <= 12.5:
        index = y_list.index(item)
median = x_list[index]
print("The estimated median is = " + str(median) + "mm")
```

```python
for item in y_list:
    if item <= 18.75:
        u_index = y_list.index(item)
upper_quartile = x_list[u_index]

for item in y_list:
    if item <= 6.25:
        l_index = y_list.index(item)
lower_quartile = x_list[l_index]

intq_range = upper_quartile - lower_quartile
print("The estimated interquartile range is = " + str(intq_range) + "mm")
```

```python
for item in x_list:
    if item <= 28:
        index = x_list.index(item)
value = y_list[index]
proportion = value / 25
print("The proportion of the strata less than 28mm thick is = " + str(proportion))
```
