| Length of life (h) | Frequency | Cumulative Frequency |
| --- | --- | --- |
| 650 &le; h < 670 | 3 | 3 |
| 670 &le; h < 680 | 7 | 10 |
| 680 &le; h < 690 | 20 | 30 |
| 690 &le; h < 700 | 17 | 37 |
| 700 &le; h < 702 | 3 | 40 |

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

x = [650, 670, 680, 690, 700, 702]
y = [0, 3, 10, 30, 37, 40]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Find the x value that corresponds to a given y value
x_list = x_new.tolist()
y_list = y_new.tolist()

median_index = int(len(y_list) / 2)
for item in y_list:
    if item <= median_index:
        index = y_list.index(item)
median = x_list[index]
print("(a) Based on a linear interpolation between the points, the median is = " + str(median))
```

```python
upper_quartile_index = int(len(y_list) * 0.75)
for item in y_list:
    if item <= upper_quartile_index:
        upper_index = y_list.index(item)
upper_quartile = x_list[upper_index]

lower_quartile_index = int(len(x_list) * 0.25)
for item in y_list:
    if item <= lower_quartile_index:
        lower_index = y_list.index(item)
lower_quartile = x_list[lower_index]

interquartile_range = upper_quartile - lower_quartile

print("(b) The interquartile range is = " + str(interquartile_range))
```
