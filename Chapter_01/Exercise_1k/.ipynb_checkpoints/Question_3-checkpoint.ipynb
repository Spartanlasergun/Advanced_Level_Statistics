```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# Define the data
x = [50, 54, 58, 62, 66, 70, 74]
y = [3, 5, 10, 22, 32, 38, 40]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Find the x value that corresponds to a given y value
x_list = x_new.tolist()
y_list = y_new.tolist()

for eggs in y_list:
    if eggs <= 20:
        index = y_list.index(eggs)
median = x_list[index]

print("Based on the curve generated, the median is = " + str(median))
```
