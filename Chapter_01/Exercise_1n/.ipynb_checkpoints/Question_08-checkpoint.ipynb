```python
import stemgraphic

stem_data = [6.60, 3.40, 6.45, 5.20, 3.60,
            7.25, 9.60, 3.75, 4.20, 8.75,
            5.75, 4.50, 3.95, 4.75, 12.25]

stemgraphic.stem_graphic(stem_data, scale=1)
```

```python
import statistics

median = statistics.median(stem_data)
data_range = max(stem_data) - min(stem_data)

print("median = " + str(median))
print("range = " + str(data_range))
```

```python
mean = statistics.mean(stem_data)
std = statistics.pstdev(stem_data)
print("mean = " + str(mean))
print("standard deviation = " + str(std))
```

```python
flat_rate = []
for item in stem_data:
    flat_increase = item + 0.30
    flat_rate.append(flat_increase)

mean = statistics.mean(flat_rate)
std = statistics.pstdev(flat_rate)
print("Under Scheme A:")
print("mean = " + str(mean))
print("standard deviation = " + str(std))
```

```python
percent_inc = []
for item in stem_data:
    percent_increase = (item * 0.05) + item
    percent_inc.append(percent_increase)

mean = statistics.mean(percent_inc)
std = statistics.pstdev(percent_inc)
print("Under Scheme B:")
print("mean = " + str(mean))
print("standard deviation = " + str(std))
```

Based on the changes in the mean, management would pay the same gross wage increase for either. However, based on the standard deviation, under scheme B the workers that have a higher wage would receive higher increase, while those with lower wages would receive less - whereas, under scheme A all workers receive the same flat pay raise and their is no inequality.
