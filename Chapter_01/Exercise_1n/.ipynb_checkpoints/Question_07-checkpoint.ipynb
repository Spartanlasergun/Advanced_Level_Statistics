```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import numpy as np

intervals = [0.78, 1.03, 1.23, 1.38, 1.58, 1.855]
cum_freq = [11, 39, 77, 111, 138, 150]

x = np.linspace(0.78, 1.855, 100)
function = interp1d(intervals, cum_freq, kind='quadratic')
y = function(x)

plt.plot(x, y)
plt.show()
```

```python
x_list = x.tolist()
y_list = y.tolist()

for item in y_list:
    if item <= 75:
        median_index = y_list.index(item)
median = x_list[median_index]

print("the esitmated median = " + str(median))
```
