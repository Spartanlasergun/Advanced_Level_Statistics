# Question Fourteen

| Cost | Number of Houses | Cum. Frequency |
| --- | --- | --- |
| 20001-50000 | 540 | 540 |
| 50001-60000 | 1150 | 1690 |
| 60001-70000 | 1320 | 3010 |
| 70001-100000 | 860 | 3870 |
| 100001-150000 | 450 | 4320 |

```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import numpy as np

intervals = [20000, 50000, 60000, 70000, 100000, 150000]
cum_freq = [0, 540, 1690, 3010, 3870, 4320]

x = np.linspace(20000, 150000, 100)
function = interp1d(intervals, cum_freq, kind='quadratic')
y = function(x)

plt.plot(x, y)
plt.show()
```

```python
x_list = x.tolist()
y_list = y.tolist()

for item in y_list:
    if item <= (4320/2):
        median_index = y_list.index(item)
median = x_list[median_index]
print("median = " + str(median))

for item in y_list:
    if item <= (4320 * 0.25):
        lower_index = y_list.index(item)
lower_quartile = x_list[lower_index]

for item in y_list:
    if item <= (4320 * 0.75):
        upper_index = y_list.index(item)
upper_quartile = x_list[upper_index]

intq_range = upper_quartile - lower_quartile
print("interquartile range = " + str(intq_range))
```
