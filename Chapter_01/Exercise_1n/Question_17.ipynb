(a) median = 8 minutes

upper quartile = 14.5
lower quartile = 5
interquartile range = 14.5 - 5 = 9.5 minutes

```python
import matplotlib.pyplot as plt

midpoints = [2.5, 7.5, 12.5, 17.5, 22.5, 27.5]
freq = [40, 66, 28, 20, 12, 4]
raw_data = []
index = 0

for item in freq:
    count = item
    while count != 0:
        raw_data.append(midpoints[index])
        count = count - 1
    index = index + 1

plt.hist(raw_data, bins=[0, 5, 10, 15, 20, 25, 30], density=True, edgecolor='black')
plt.show()
```

```python
import statistics

mean = statistics.mean(raw_data)
print("estimated mean = " + str(mean))
```
