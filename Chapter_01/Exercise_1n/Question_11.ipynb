# Question Eleven

|  | Number of Candidates | Mean Mark | Variance |
| --- | --- | --- | --- |
| Set A | 20 | 66 | 9 |
| Set B | 30 | 51 | 39 |

$$ \bar x = \frac{\Sigma x}{n} $$

hence,

$$ \Sigma x = \bar x * n $$

$$ \Sigma x_1 = 20 * 66 = 1320 $$

$$ \Sigma x_2 = 30 * 51 = 1530 $$

$$ \bar x_{1+2} = \frac{\Sigma x_1 + \Sigma x_2}{n_1 + n_2} $$

$$ \bar x_{1+2} = \frac{1320 + 1530}{20 + 30} = \frac{2850}{50} $$

$$ \bar x_{1+2} = 57 $$

-----

$$ \sigma^2 = \frac{\Sigma x^2}{n} - \bar x^2 $$

hence,

$$ \Sigma x^2 = n(\sigma^2 + \bar x^2) $$

$$ \Sigma x^2_1 = 20(9 + 66^2) = 87300 $$

$$ \Sigma x^2_2 = 30(39 + 51^2) = 79200  $$

Thus,

$$ \sigma _{1+2} = \sqrt{\frac{\Sigma x^2_1 + \Sigma x^2_2}{n_1 + n_2} - \bar x^2_{1+2}} $$

$$ \sigma _{1+2} = \sqrt{\frac{87300 + 79200}{20 + 30} - (57^2)} $$

$$ \sigma _{1+2} = \sqrt{3330 - 3249} = 9 $$
