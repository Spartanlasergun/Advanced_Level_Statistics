| --- | --- | --- |
| &le; 1 | 20 | 20 |
| &le; 2 | 67 | 87 |
| &le; 2$\frac{1}{2}$ | 118 | 205 |
| &le; 3 | 177 | 382 |
| &le; 5 | 315 | 697 |
| &le; 10 | 400 | 1097 |


```python
import matplotlib.pyplot as plt

midpoints = [0.5, 1.5, 2.25, 2.75, 4, 7.5]
freq = [20, 67, 118, 177, 315, 400]
raw_data = []
index = 0

for item in freq:
    count = item
    while count != 0:
        raw_data.append(midpoints[index])
        count = count - 1
    index = index + 1

plt.hist(raw_data, bins=[0, 1, 2, 2.5, 3, 5, 10], density=True, edgecolor='black')
plt.show()
```

```python
from scipy.interpolate import interp1d
import numpy as np

intervals = [0, 1, 2, 2.5, 3, 5, 10]
cum_freq = [0, 20, 87, 205, 382, 697, 1097]

x = np.linspace(0, 10, 100)
function = interp1d(intervals, cum_freq, kind='quadratic')
y = function(x)

plt.plot(x, y)
plt.show()
```

```python
x_list = x.tolist()
y_list = y.tolist()

for item in y_list:
    if item <= (1097/2):
        median_index = y_list.index(item)
median = x_list[median_index]
print("estimated median = " + str(median))
```
