| Number Of Goals | Frequency | Cumulative Frequency |
| --- | --- | --- |
| 0 | 0 | 0 |
| 1 | 1 | 1 |
| 2 | 3 | 4 |
| 3 | 2 | 6 |
| 4 | 5 | 11 |
| 5 | 8 | 19 |
| 6 | 6 | 25 |



```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 1, 4, 6, 11, 19, 25]

plt.step(x, y, where='post')
plt.title("Step Diagram")
plt.xlabel("Number of Goals")
plt.ylabel("Cumulative Frequency")
plt.show()
```

```python
import statistics

freq = [0, 1, 3, 2, 5, 8, 6]
raw_data = []
index = 0
for item in freq:
    count = item
    while count != 0:
        raw_data.append(x[index])
        count = count - 1
    index = index + 1
median = statistics.median(raw_data)
quartiles = statistics.quantiles(raw_data, n=4)
intq_range = quartiles[2] - quartiles[0]

print("c) Median = " + str(median))
print("d) Interquartile Range = " + str(intq_range))
```
