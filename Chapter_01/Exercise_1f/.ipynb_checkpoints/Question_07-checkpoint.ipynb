```python
import statistics
import matplotlib.pyplot as plt

data = [0.5, 1.5, 2.5, 4, 7.5, 10.5]
freq = [6, 10, 15, 5, 4, 0]
raw_data = []
index = 0
for item in freq:
    count = item
    while count != 0:
        raw_data.append(data[index])
        count = count - 1
    index = index + 1
mean = sum(raw_data) / len(raw_data)
std = statistics.pstdev(raw_data)
print("The mean is = " + str(mean) + " minutes")
print("The standard deviation is = " + str(std))

plt.hist(raw_data, bins=[0, 1, 2, 3, 5, 10, 12], density=True, edgecolor="black")
plt.show()
```
