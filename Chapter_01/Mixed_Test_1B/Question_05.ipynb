```python
import stemgraphic

data = [12, 20, 14, 17, 17, 8, 19, 13, 27, 13,
        16, 18, 10, 7, 22, 16, 11, 18, 13, 6,
        16, 12, 14, 23, 15, 8, 10, 17, 16, 19]

stemgraphic.stem_graphic(data, scale=10)
```

```python
import statistics

median = statistics.median(data)
print("median = " + str(median))

quartiles = statistics.quantiles(data, n=4)
print("lower quartile = " + str(quartiles[0]))
print("upper quartile = " + str(quartiles[2]))
```

```python
import seaborn as sns

sns.boxplot([min(data), quartiles[0], median, quartiles[2], max(data)], orient='h', whis=max(data))
```
