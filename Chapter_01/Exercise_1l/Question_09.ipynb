```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# Define the data
x = [3, 5, 7.5, 9, 10, 11.5, 14, 16]
y = [6, 21, 48, 97, 149, 178, 191, 200]

f = interp1d(x, y, kind='quadratic')

# Generate new x and y values for the curve
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x_new, y_new)

# Plot the original data points on top of the interpolated curve
plt.scatter(x, y)

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Find the x value that corresponds to a given y value
x_list = x_new.tolist()
y_list = y_new.tolist()

median_freq = 100
for item in y_list:
    if item <= median_freq:
        median_index = y_list.index(item)
median = x_list[median_index]

print("(a)")
print("The estimated median is = " + str(median))

for item in y_list:
    if item <= 150:
        upper_index = y_list.index(item)
upper_quartile = x_list[upper_index]

for item in y_list:
    if item <= 50:
        lower_index = y_list.index(item)
lower_quartile = x_list[lower_index]

intq_range = upper_quartile - lower_quartile
semi_intq_range = intq_range / 2

print("The estimated semi interquartile range of the data is = " + str(semi_intq_range))
```

```python
std = 3.03
ratio = semi_intq_range / std
print("(b) The ratio of the semi-interquartile range to the standard deviation is = " + str(ratio))
```
