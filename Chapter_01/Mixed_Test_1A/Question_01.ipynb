| --- | --- | --- |
| 65 | 0 | 0 |
| 85 | 25 | 25 |
| 95 | 28 | 53 |
| 105 | 20 | 73 |
| 115 | 17 | 90 |
| 155 | 10 | 100 |

```python
import matplotlib.pyplot as plt
import statistics

freq = [25, 28, 20, 17, 10]
midpoints = [75, 90, 100, 110, 135]
raw_data = []
index = 0

for item in freq:
    count = item
    while count != 0:
        raw_data.append(midpoints[index])
        count = count - 1
    index = index + 1

plt.hist(raw_data, bins=[65, 85, 95, 105, 115, 155], density=True, edgecolor='black')
plt.show()
```

```python
from scipy.interpolate import interp1d
import numpy as np

cum_freq = [0, 25, 53, 73, 90, 100]
boundaries = [0, 85, 95, 105, 115, 155]

x = np.linspace(0, 155, 100)
function = interp1d(boundaries, cum_freq, kind='quadratic')
y = function(x)

x_list = x.tolist()
y_list = y.tolist()

for item in y_list:
    if item <= 50:
        est_mean_index = y_list.index(item)
est_mean = x_list[est_mean_index]
print("estimated mean = " + str(est_mean))
```
