| T (minutes) | N (runners) | Cum. Frequency |
| --- | --- | --- |
| 65 | 0 | 0 |
| 85 | 25 | 25 |
| 95 | 53 | 78 |
| 105 | 73 | 151 |
| 115 | 90 | 241 |
| 155 | 100 | 341 |

```python
import matplotlib.pyplot as plt
import statistics

freq = [0, 25, 53, 73, 90, 100]
midpoints = [32.5, 75, 100, 110, 115, 135]
raw_data = []
index = 0

for item in freq:
    count = item
    while count != 0:
        raw_data.append(midpoints[index])
        count = count - 1
    index = index + 1

plt.hist(raw_data, bins=[0, 65, 85, 95, 105, 115, 155], edgecolor='black')
plt.show()
```

```python
from scipy.interpolate import interp1d
import numpy as np

cum_freq = [0, 25, 78, 151, 241, 341]
boundaries = [0, 85, 95, 105, 115, 155]

x = np.linspace(0, 155, 100)
function = interp1d(boundaries, cum_freq, kind='quadratic')
y = function(x)

x_list = x.tolist()
y_list = y.tolist()

for item in y_list:
    if item <= (341/2):
        est_mean_index = y_list.index(item)
est_mean = x_list[est_mean_index]
print("estimated mean = " + str(est_mean))
```
