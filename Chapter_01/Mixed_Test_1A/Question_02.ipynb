(a) The totals represent the number of datapoints in each grouping. The missing totals are: 7, 6, 4, and 8

```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import numpy as np

freq = [12, 9, 10, 7, 6, 4, 8, 3, 4, 3]
cum_freq = [12, 21, 31, 37, 43, 47, 55, 58, 62, 65]
boundaries = [5.5, 6.0, 6.5, 7.0, 7.5, 8.0, 8.5, 9.0, 9.5, 10]

x = np.linspace(5.5, 10, 100)
function = interp1d(boundaries, cum_freq, kind='quadratic')
y = function(x)

plt.plot(x, y)
plt.show()
```

```python
x_list = x.tolist()
y_list = y.tolist()

for item in y_list:
    if item <= (65 * 0.5):
        median_index = y_list.index(item)
median = x_list[median_index]

print("estimated median = " + str(median))

for item in y_list:
    if item <= (65 * 0.25):
        lower_index = y_list.index(item)
lower_quartile = x_list[lower_index]

print("estimated lower quartile = " + str(lower_quartile))

for item in y_list:
    if item <= (65 * 0.75):
        upper_index = y_list.index(item)
upper_quartile = x_list[upper_index]

print("estimated upper quartile = " + str(upper_quartile))
```

```python
import seaborn as sns

sns.boxplot([5.0, lower_quartile, median, upper_quartile, 9.9], orient='h', whis=10)
```

The distribution is positively skewed
