```python
x = [0, 1, 2, 3, 4]
freq = [4, 12, 6, 2, 1]
raw_data = []
index = 0
for item in freq:
    count = item
    while count != 0:
        raw_data.append(x[index])
        count = count - 1
    index = index + 1

import matplotlib.pyplot as plt
import seaborn as sns

plt.title("(a)")
sns.boxplot(x=raw_data)
```
