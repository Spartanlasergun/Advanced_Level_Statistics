```python
data = [67, 76, 85, 42, 93, 48, 93, 46, 52,
        72, 77, 53, 41, 48, 86, 78, 56, 80,
        70, 70, 66, 62, 54, 85, 60, 58, 43,
        58, 74, 44, 52, 74, 52, 82, 78, 47,
        66, 50, 67, 87, 78, 86, 94, 63, 72,
        63, 44, 47, 57, 68, 81]

import statistics

quantiles = statistics.quantiles(data, n=4)

print("median = " + str(quantiles[1]))
print("upper quartile = " + str(quantiles[2]))
print("lower quartile = " + str(quantiles[0]))
```

```python
import seaborn as sns

sns.boxplot(x=data, whis=max(data))
```
