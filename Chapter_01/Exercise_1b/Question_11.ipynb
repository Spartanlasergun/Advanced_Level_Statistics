```python
import matplotlib.pyplot as plt

midpoints = [20, 35, 45, 55, 62, 67, 77]
frequencies = [22, 18, 22, 24, 14, 12, 24]
raw_data = []
index = 0
for item in frequencies:
    count = item
    while count != 0:
        raw_data.append(midpoints[index])
        count = count - 1
    index = index + 1

plt.hist(raw_data, bins=[9.5, 29.5, 39.5, 49.5, 59.5, 64.5, 69.5, 84.5], density=True, edgecolor="black")
```
