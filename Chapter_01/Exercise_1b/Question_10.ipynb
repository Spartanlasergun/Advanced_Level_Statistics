```python
import matplotlib.pyplot as plt

data = [5, 12, 17, 24, 34, 45]
frequencies = [5, 8, 32, 41, 15, 2]
raw_data = []
index = 0
for item in frequencies:
    count = item
    while count != 0:
        raw_data.append(data[index])
        count = count - 1
    index = index + 1

plt.hist(raw_data, bins=[0, 9.5, 14.5, 19.5, 29.5, 39.5, 50], density=True, edgecolor="black")
plt.show()
```
