```python
import matplotlib.pyplot as plt

time = [5, 10, 20, 25, 40, 45]
frequency = [2, 12, 7, 15, 2, 0]
raw_data = []
count = 0
for data in frequency:
    unit = data
    while unit != 0:
        raw_data.append(time[count])
        unit = unit - 1
    count = count + 1
print(raw_data)
plt.hist(raw_data, bins=[5, 10, 20, 25, 40, 45], density=True, edgecolor='black')
plt.show()
```
