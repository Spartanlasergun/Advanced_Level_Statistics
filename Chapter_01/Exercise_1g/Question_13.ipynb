```python
mean = sum(data) / len(data)
print("The mean age is = " + str(mean))
```

```python
sum_x = 37.72 * 18
# sum_x + sum_y = mean_xy * 30
sum_y = (mean * 30) - sum_x
mean_y = sum_y / 12
print("The mean age of the women is = " + str(mean_y))
```
